```python
#Q1. A python program to create lists with different types of elements

#a general way to create lists
# Method 1: create a list with integer numbers
num = [ 10,20,30,40,50]
print("Total list: ",num) # display total list
print("First Element:%d , Last Element:%d" %(num[0],num[4])) #Display first and last elements
print("Type of list of elements",type(num[0]))

# Method 2: create a list with strings
names = ["Radha","Jay","Priya","Aditya","Om"]
print("Total list: ",names) # display total list
print("First Element:%s , Last Element:%s" %(names[0],names[4])) #Display first and last elements
print("Type of list of elements",type(names[0]))

```

```python
#Q2. A python program to create lists using range function
# creating lits using range() function
# create a list with 0 to 9 consecutive integer numbers
list1 = range(10)
for i in list1: # display elemnts by element
    print(i,', ',end='')
print()# throw cursor to next line

#create list with integers from 5 to 9
list2 = range(5,10)
for i in list2:
    print(i,', ', end ='')
print()

#create list with odd numbers from 5 to 9
list3 = range(5,10,2)
for i in list3:
    print(i,', ', end ='')
print()

```

```python
#Q3 A python program to access list elements using loops
#displaying list elemnts using while and for loops
list = [10,20,30,40,50]
print("using while loop")
i =0
while i < len(list): #repeat from 0 to length of first
    print(list[i])
    i= i+1
print('using for loop')
for i in list: # repeat for all elements
    print(i)
```

```python
#Q4 A python program to display the elements of a list in reverse order.
days = ['Sunday','Monday','Tuesday','Wednesday','Thrusday','Friday']

print("\nIn reverse order: ")
i = len(days)-1 # i will be 4
while i>=0:
    print(days[i]) # display from 4th to 0th elements
    i-=1
print('\nIn reverse order: ')
i=-1 #days[-1] represents last element
while i>=-len(days): # display from -1th to -5th elements
    print(days[i])
    i-=1
```

```python
#Q5. A python program to concatenation of Two lists and repetition of lists
x = [10,20,30,40,50]
y = [100,200,300,400]
print("concatenated list: ",x+y) #concatenate x and y

# repetition of list
print("repetition of list:",x*2) # repeat the list x for 2 times
```

```python
#Q6. A python program to understand list processing methods
# python's list methods
num = [1,2,3,4,5]

n = len(num)
print('No. of elements in num: ',n)

print('sum of num',sum(num))
num.append(6)
print('num after appending 6: ',num)
num.insert(0,50)
print('num after inserting 5 at 0th position: ', num)
num1 = num.copy()
print('Newly created list num1: ',num1)
num.extend(num1)
print('num after appending num1: ',num)

n = num.count(5)
print('No. of 5 found in the list num: ', n)
num.pop()
print('num after removing ending elements: ',num)
num.sort()
print('num after sorting: ',num)
num.reverse()
print('num after reversing: ',num)
num.clear()
print('num after removing all elements: ',num)




```
